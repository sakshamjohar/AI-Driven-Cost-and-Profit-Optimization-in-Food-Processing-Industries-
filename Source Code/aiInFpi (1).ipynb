```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (or create a synthetic one if you haven't already)
# df = pd.read_csv("food_processing_data.csv")  # remove this line if you want to generate synthetic data

# If you want to generate a quick synthetic version instead:
# (only keep this block if you remove the df = pd.read_csv(...) line)
import numpy as np
np.random.seed(42)
n = 100
raw = np.random.uniform(8000, 15000, n)
energy = np.random.uniform(1200, 2500, n)
labor = np.random.uniform(2000, 4000, n)
maint = np.random.uniform(500, 1500, n)
qty = np.random.randint(500, 1200, n)

total_cost = raw + energy + labor + maint
revenue = qty * np.random.uniform(25, 35) + np.random.uniform(5000, 10000, n)
profit = revenue - total_cost

df = pd.DataFrame({
    "total_cost": total_cost,
    "profit": profit.clip(min=0)
})

# --- Create the scatter plot (Fig. 1) ---
plt.figure(figsize=(7, 5))
plt.scatter(df["total_cost"], df["profit"], alpha=0.7, color="blue", edgecolors="black", s=40)
plt.xlabel("Total Cost per Batch (₹)")
plt.ylabel("Profit per Batch (₹)")
plt.title("Cost vs Profit in Food Processing Unit")
plt.grid(True, alpha=0.3)

# Save the figure (this matches your IEEE caption)
plt.tight_layout()
plt.savefig("fig1_cost_vs_profit.png", dpi=300, bbox_inches="tight")
plt.show()  # only needed in notebooks or IDEs
```
